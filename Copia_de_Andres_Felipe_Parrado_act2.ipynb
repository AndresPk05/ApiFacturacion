Un modelo de aprendizaje que utiliza el gradiante descendiente es el de regresion lineal, este se utiliza para poder ajustar el parametros de aprendizaje del modelo, tanto la taza de aprendizaje como la cantidad de iterazaciones a realizar para encontrar la combinacion de parametros optimos y reducir el error.

El algoritmo que se implementara acontinuacion fue extraido de https://github.com/albgarse/Artificial-Intelligence/blob/master/Machine%20Learning/Regresion%20lineal.ipynb, se modifica la tasa de aprendizaje y las iteraciones para visualizar como cambiaria la regresion linea dependiendo los parametros de entrada.

```python
import pandas as pd
import random
import matplotlib.pyplot as plt

def gen_data(n, bias, varianza):
  x = []
  y = []
  for i in range(0, n):
    x.append(i)
    y.append((i + bias) + random.uniform(0,1) * varianza)
  return x, y

x, y = gen_data(100, 25, 50)
plt.scatter(x, y)
```

```python
def coste(x, y, a, b):
  m = len(x)
  error = 0.0
  for i in range(m):
    hipotesis = a + b * x[i]
    error += (y[i] - hipotesis) ** 2
  return error / (2*m)


def descenso_gradiante(x, y, a, b, alpha, epochs):
  m = len(x)
  hist_coste = []
  for ep in range(epochs):
    b_deriv = 0
    a_deriv = 0
    for i in range(m):
      hipotesis = a + b * x[i]
      a_deriv += hipotesis - y[i]
      b_deriv += (hipotesis - y[i]) * x[i]
      hist_coste.append(coste(x, y, a, b))
    a -= (a_deriv / m) * alpha
    b -= (b_deriv / m) * alpha

  return a, b, hist_coste
```

```python
a =1
b =1
alpha = 0.0005
iters = 20000
a, b, hist_coste = descenso_gradiante(x, y, a, b, alpha, iters)
```

```python
plt.scatter(x, y)
pred_x = [0, max(x)]
pred_y = [a + b * 0, a + b * max(x)]
plt.title('20000 Iteraciones')
plt.axis([0, 100, 0, 200])
plt.plot(pred_x, pred_y, "r")
```

```python
x_base = range(len(hist_coste))
plt.plot(x_base[1000:], hist_coste[1000:])
```
